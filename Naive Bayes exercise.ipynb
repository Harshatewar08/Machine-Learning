```python
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```python
df
```

```python
df['target'] = wine.target
df
```

```python
df[50:70]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(wine.data,wine.target, test_size=0.18) 
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
mg = GaussianNB()
```

```python
mg.fit(X_train,y_train)
```

```python
mg.score(X_test,y_test)
```

### When the test size of the model is 15%, the score for the GaussianNB model is 100%

```python
mn = MultinomialNB()
```

```python
mn.fit(X_train,y_train)
```

```python
mn.score(X_test,y_test)
```

### ### When the test size of the model is 18%, the score for the multinomialNB model is 88%
